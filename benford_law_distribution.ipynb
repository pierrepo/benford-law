# Repr√©sentation de la loi de Benford

```python
import benford as ben
import matplotlib.pyplot as plt
```

## Loi de benford pour le $1^{er}$ chiffre

```python
plt.figure(figsize=(8, 6))
plt.rc('font', size=15)
plt.bar(range(1,10), ben.get_theoretical_freq_benford())
plt.xlabel("First digit")
plt.ylabel("Proportion")
plt.title("Theoretical distribution of Benford law in base 10")
plt.xticks(range(1, 10))
plt.savefig("Proportion_Benford_law_1digit.png")
```

## Loi de benford pour les deux $1^{er}$ chiffres

```python
plt.rc('font', size=40)
plt.figure(figsize=(40,15))
plt.bar(range(10,100), ben.get_theoretical_freq_benford(2))
plt.xlabel("Two first digit")
plt.ylabel("Proportion")
plt.title("Theoretical Benford law")
plt.xticks(range(10, 100, 10), labels=("10", "20", "30","40", "50", "60", "70", "80", "90"));
plt.savefig("Proportion_Benford_law_2digit.png")
```
